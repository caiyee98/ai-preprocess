# Rename multiple files

```python
import os

# Rename multiple files
def main():
   i =0
   path=r"path/to/dir/" 
   for filename in os.listdir(path):
      my_dest =str(i) +"_ct" + ".jpg"
      my_source =path + filename
      my_dest =path + my_dest
      os.rename(my_source, my_dest)
      i += 1

main()
```

```python

from PIL import Image
import os

path = "path/to/dir/"
dirs = os.listdir( path )

def resize():
    for item in dirs:
        if os.path.isfile(path+item):
            im = Image.open(path+item)
            f, e = os.path.splitext(path+item)
            imResize = im.resize((64,64), Image.ANTIALIAS)
            imResize.save(f + ' resized.jpg', 'JPEG', quality=90)

resize()
```

# Resize and pad with Image module Using PIL

```python
from PIL import Image, ImageOps
import matplotlib.pyplot as plt
import imageio as ia
import os

def main(): 
	inPath ="path/to/input_dir"
	outPath ="path/to/output_dir"

	for imagePath in os.listdir(inPath): 
		inputPath = os.path.join(inPath, imagePath) 

		desired_size = 64
		# im_pth = "C:/Users/EndUser/Desktop/New folder/2020.02.25.20021568-p23-108_0.png"
		im = Image.open(inputPath)
		old_size = im.size  # old_size[0] is in (width, height) format

		ratio = float(desired_size)/max(old_size)
		new_size = tuple([int(x*ratio) for x in old_size])

		im = im.resize(new_size, Image.ANTIALIAS)
		# create a new image and paste the resized on it

		new_im = Image.new("RGB", (desired_size, desired_size))
		new_im.paste(im, ((desired_size-new_size[0])//2,(desired_size-new_size[1])//2))

		fullOutPath = os.path.join(outPath, imagePath) 
		new_im.save(fullOutPath)

main() 
# new_im.show()
# plt.imshow(new_im)
```

# Save files in a folder to a text file

```python
import os

outputfile	= "path/to/dir"  # file to save the results to
folder		= "path"		# the folder of images name to be saved
exclude		= ['Thumbs.db','.tmp']	# exclude files containing these strings
pathsep		= " / "

with open(outputfile, "w") as txtfile:
	for path,dirs,files in os.walk(folder):
		sep = "\n---------- " + path.split(pathsep)[len(path.split(pathsep))-1] + " ----------"

		for fn in sorted(files):
			if not any(x in fn for x in exclude):
				filename = os.path.splitext(fn)[0]
				
				print (filename +".png")
				txtfile.write( filename +".png" +"\n" )

txtfile.close()
```
